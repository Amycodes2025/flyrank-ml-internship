## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

For my lane, I am mixing two types of ML tasks. The first part is classification, I'm trying to guess if a page is declining or not just like the label we used in the starter notebooks. The second part is ranking, once I know which pages mught be declining, I still need to put them in order so the editor knows which one to look at first. They can't check every page so ranking helps them focus on the most important ones first. Ranking helps them focus on the most important ones first. Really, the whole output is decision-support, it doesn't replace the editor's judgment, it just helps them prioritize where to look first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



What I'd predict is wheather a page is declining or not, using the same label the starter notebooks used (is_declining_label) which comes from trend_direction. This isn't an observed future outcome, it's a defined rule, the label is just calculated from the current data not something that actually happened later in time. A stronger version would predict a real future outcome instead like using the last 90 days of data to predict if a page will decline in the next 30 days. That would be a proper observed outcome instead of a snapshot rule.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The metric I'd use is Precision@50. This means that out of the top 50 pages I flag as needing review, how many actually turn out to be true positives (really declining)? I'm choosing this over plain accuracy because an editor can only review a limited number of pages at a time around 50 so what matters isn't how well the model does across all 30,000 pages, it's how good the top of the list is since that's the only part anyone actually acts on. A good number would be well above the baseline hand rule's 0.240, the random forest in the starter pipeline measured 0.740 Precision@50 which is the kind of result that would actually be worth building on.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple if-statement rule like 'if a page hasn't been updated in 180 days and still gets traffic, flag it' only looks at two things at a time. But real pages are messier than that, a page could be old but still doing fine or fairly new but already dropping fast. There are too many small factors mixing together (traffic, position, click rate, word count, age) for one simple rule to catch all the patterns. In Notebook 02, I saw this myself: my own hand written rule got some pages right but a decision tree which can weigh several signals together instead of just one or two  did a better job (0.720 vs 0.600 Precision@50 when I went from a depth-2 to depth-3 tree). And in the full pipeline, a random forest model beat the hand rule by a big margin (0.740 vs 0.240 Precision@50). That's strong evidence though not proof that a learned pattern can catch more cases than a fixed if-statement, the comparison is directional based on this specific dataset, not a guarantee it'll hold everywhere."
